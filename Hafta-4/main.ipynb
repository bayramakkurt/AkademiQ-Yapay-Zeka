## CSV Veri Okuma

```python
df2 = pd.read_csv("Chocolate Sales.csv")
df2   #Büyük verilerde ilk 5 ve son 5 veriyi yazdırır.
```

```python
df2.head()  #İlk 5 veriyi getirir
```

```python
df2.tail() #Son 5 veriyi getirir.
```

```python
df2["Amount"] #Sadece istenen sütun bilgilerini getirir.
```

```python
df2[ df2["Boxes Shipped"] > 300 ] #İstenen koşula göre filtreleme yapar filtreye uygun elemanları sadece getirir.
```

```python
df2[ df2["Boxes Shipped"] > 300  ]["Amount"] #Bi şarta uyan elemanların sadece belirli sütununu yazdırmak için [] [] kullanılır.
```

```python
df2.isnull() #Bu komut her veri kutucuğu için alandaki değerlerin Null olup olmadığını gösterir.
```

```python
df2.isnull().any()  #Sütunların tamamında herhangi bir null değer varmı yokmu onu sütun sütun gösterir.
```

```python
df2.isnull().sum() #Null değerlerin sayısını dönderir.
```

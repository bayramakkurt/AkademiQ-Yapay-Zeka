```python
import nltk
nltk.download("stopwords")
```

```python
# hediye HeDiYe HEDiye HEDİYE H3DİYE gibi metinde geçen kelimeleri belirli işlemlerle metinden sadece kullanacağımız kelimeleri çıkarmayı hedefliyoruz.
#Makine öğrenmesinde metin işlenmez metinleri sayısal değerlere çevirmemiz lazım.

#1.Aşama => Metni sabit formda küçük ve noktalama işaretsiz hale getirdik.
#2.Aşama => Metni kelimelerine ayırdık.
#3.Aşama => Kelimeler içinde sayısal olan değerleri kaldırdık.
#STEM => Kelimeleri kökenine indirgemektir.Yani kelimelerin hepsini kökenine indirger.
#REMOVE STOPWORDS => Ne,ve,veya,ki,nasıl,gibi,the,is,at,on gibi cümlenin ana bağlamını değiştirmeyen kelimeleri ayıklar.

import string

ps = nltk.stem.porter.PorterStemmer() # Kök bulucu işlemi için yani STEM yapacak nesneyi oluşturur.

def clean_text(text):
    text = text.lower() #Bütün harfleri  küçülttük.

    text = text.translate( str.maketrans( '','',string.punctuation ) )  #Buradaki amacımız string.punctuation(noktalama işaretlerini içeren değişken) değişkeni ile bütün noktalama işaretleri yerine boş karakter eklemek.

    words = text.split() #text değişkenindeki her bir kelimeyi alıp words değişkeninde liste olarak tutar. Parantez içine yazılan parametreler arasında kalan kelimeleri alır default boşluk olarak işlem yapar.

    words = [ word for word in words if not word.isdigit() ] #Buradaki işlem words içindeki her kelimeyi tek tek for ile gezer ve if koşuluyla isdigit() sayısal olan verileri almayan sadece metinsel verileri içinde barındıracak şekilde bıraktı.

    words = [ ps.stem(word) for word in words if word not in nltk.corpus.stopwords.words('english') ] #Bütün kelimeleri gezip STOPWORDS kelimeleri arasında değilse listede tutmak için kullandık.Ek olarakta kelimelerin kök haline çevirip tutar.

    return " ".join(words) #Amacımız doğrultusunda elde ettiğimiz kelimeleri boş stringe ekler ve return eder.
```

### Elimizdeki dataframe içindeki text sütununda bulunan metni fonksiyona gönderip dönen sonucu clean_text adında yeni bir sütun açıp oraya kaydeder.

```python
df["clean_text"] = df["text"].apply(clean_text)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X = df["clean_text"]
y = df["label_num"]
```

### Metinsel olan temizlenmiş metinlerimizi vektörel değere çevirip daha hızlı eğitilecek hale getirmemiz gerek.<br>
### VECTORİZE => Metni numerize etmemize yarar.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform( df["clean_text"] )
X_tfidf
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_tfidf, y, test_size=0.3, random_state=42 )
```

```python
#Naive Bayes Algoritması => Bir değerin bir olasılığı nasıl etkiledğini Bayes Teoremi 
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
```

```python
model.fit( X_train, y_train )
```

```python
y_pred = model.predict( X_test )
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print( classification_report( y_test, y_pred ) )
```

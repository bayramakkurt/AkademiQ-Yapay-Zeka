```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("HousingData_Clean.csv" , index_col=0)  #index_col parametresi baştaki index sütununu  okunan dataframe sütun olarak dahil etmez.
df.head()
```

# DECİSİON TREE (KARAR AĞAÇLARI)

## Belirli bir şarta göre karar vererek recursive şekilde derinleşen şartlara göre dallanabilir.(B Tree mantığıyla aynı görsel) Max Step = Derinlik Değeri <br>


```python
X = df.drop( columns=["MEDV"] )
y = df["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42 )
```

```python
from sklearn.tree import DecisionTreeRegressor   #Fiyat tahmini yapacağımız için REgressor seçildi.
```

```python
model = DecisionTreeRegressor( max_depth=4 , random_state=42 ) #Derinlik artarsa hata oranı azalır diye bir kaide yok doğru derinlik değeri deneme ile bulunmalı.
model.fit( X_train, y_train )
```

```python
y_pred = model.predict( X_test )
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error( y_pred, y_test )
mae
```

```python
mse = mean_squared_error( y_pred, y_test )
mse
```

```python
r2 = r2_score( y_pred, y_test ) #-1 ile 1 arasında skor alır. -1 kötü , 1 iyi
r2
```

```python
#Overfitting => Ezberleme: 
```

```python
from sklearn.tree import plot_tree

plt.figure( figsize=(20,10) )
plot_tree( model, feature_names=X.columns, filled=True ) #Karar ağaçlarını görselleştirmek için kullanılır. Parametre olarak model ve sütun adlarını alır.
plt.show()                 #Filled true parametresi ise ağaçları renklendirir. Renkler koyu ise kararda emin olduğunu , açık ise kararda pek emin olmadığını belirtir..
```

```python
#GridSearchCV, RandomSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { #Max depth veri sayısı kombinasyonlarını geçmemesi önerilir.
    "max_depth": list(range(1,500))
}
```

```python
grid = GridSearchCV( DecisionTreeRegressor(), param_grid, scoring="r2", cv=10 ) #Griddeki her parametre kombinasyonu için o kombinasyonla bir model eğit ve en iyi modeli bul.
grid.fit( X_train, y_train )  #Scoring skorlamaya göre değerlendirir.cv parametresi ise modeli 
```

```python
grid.best_estimator_, grid.best_params_, grid.best_score_
```

# RANDOM FOREST

## Birden fazla Karar Ağacından oluşan yapı ve bu ağaçların ortalamasını alıp daha doğru sonuç verir.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor( n_estimators=10, random_state=42, max_depth=4 )  #n_estimator parametresi ormanın kaç karar ağacından oluşacağını belirler.
model2.fit( X_train, y_train )
```

```python
y_pred2 = model2.predict(X_test)
```

```python
mae2 = mean_absolute_error( y_pred2, y_test )
mse2 = mean_squared_error( y_pred2, y_test )
r2_2 = r2_score( y_pred2, y_test )
```

```python
( "Random Forest:", mae2, mse2, r2_2 )
```

```python
( "Decision Tree:", mae, mse, r2 )
```

```python
import pandas as pd
```

## CSV dosyasını okuma

```python
data = pd.read_csv("HousingData.csv")
data
```

```python
dataframe = pd.DataFrame(data)
dataframe
```

## Dataframe içindeki kolonların veri türlerini kontrol etme

```python
dataframe.dtypes
```

```python
dataframe.isnull().any()
```

## Dataframe içindeki boş değerlerin toplam sayısına bakma

```python
dataframe.isnull().sum()
```

## CRIM sütunundaki boş değerleri ortalama değer ile doldurma.

```python
dataframe["CRIM"].fillna( dataframe["CRIM"].mean() , inplace=True )
dataframe.isnull().sum()
```

## ZN sütunundaki boş değerleri en çok tekrar eden elemanla doldurma.

```python
dataframe["ZN"].fillna( dataframe["ZN"].mode()[0] , inplace=True )
dataframe.isnull().sum()
```

## INDUS sütunundaki boş değerleri alt ve üstteki değerlere eşit değerlerle doldurma.

```python
dataframe["INDUS"]  = dataframe["INDUS"].ffill().bfill()
dataframe.isnull().sum()
```

## Diğer bütün sütunlardaki boş değerleri ortalama değerler ile doldurma

```python
dataframe["CHAS"].fillna( dataframe["CHAS"].mean() , inplace=True )
dataframe.isnull().sum()
```

```python
dataframe["AGE"].fillna( dataframe["AGE"].mean() , inplace=True )
dataframe.isnull().sum()
```

```python
dataframe["LSTAT"].fillna( dataframe["LSTAT"].mean() , inplace=True )
dataframe.isnull().sum()
```

```python
dataframe
```

## Oluşan yeni CSV dosyasını locale kaydetme

```python
dataframe.to_csv( "New_HousingData.csv" , index=False )
```

## Oda Sayıfına Göre Ortalama Ev Fiyatları

```python
plt.figure(figsize=(8, 6))
plt.plot(dataframe.index, dataframe['CHAS'], label="Charles Nehri'ne yakınlık")
plt.plot(dataframe.index, dataframe['MEDV'], label='Ortalama Ev Fiyatı')

plt.xlabel('Index')
plt.ylabel('Değer')
plt.title('İki Sütunun Karşılaştırması')
plt.legend()
plt.grid(True)
plt.show()
```

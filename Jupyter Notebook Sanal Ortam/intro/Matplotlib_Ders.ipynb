```python
import matplotlib.pyplot as plt
```

```python
x = [ 1,2,3,4,5 ]
y = [ 100,200,300,400,500 ]
x , y
```

```python
plt.plot( x,y ,  color="Red" , linestyle = "dashed" , marker="s" , label="Satışlar")  #Grafik oluşturur. 
#Color=Çizginin rengi değiştirir.Linestyle=Çizginin türünü belirtir.Marker=Square.label=Başlık

plt.legend()  #Bilgi alanını gösterir.
plt.grid()  #Karesel alanı gösterir.
plt.xlabel("Günler") #X eksenine başlık ekler.
plt.ylabel("Satışlar") #Y eksenine başlık ekler.
plt.show() #Grafiği gösterir.
```

```python
import numpy as np
```

```python
x = np.linspace( 0,10,100 , dtype=float ) #Belirlenen aralıkta ve istenen adette sayı üretir. 0-10 arası 100 adet sayı.
x
```

```python
y1 = np.sin( x ) 
y2 = np.cos( x )
y1 , y2
```

```python
plt.figure( figsize=(8,6) )  #Boş bir sayfa açar. Altındaki bütün plotları kendisinde toplar.Figsize inch olarak boyutu belirler.
plt.plot( x,y1 , label="Sin(x)" , color="red" )
plt.plot( x,y2 , label="Cos(x)" , color="blue" )

plt.xlabel("X Değeri")
plt.ylabel("Sin-Cos Değerleri")
plt.legend()

plt.show()
```

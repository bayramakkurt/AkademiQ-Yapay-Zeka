# KNN Algoritmasını Regresyon Probleminde Kullanma

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
```

## Veri Kümelerini Oluşturma

```python
X= np.sort(5*np.random.rand(40,1), axis=0) #40 adet 0-1 arasında sayı üretir.ÜNiform dağılım. FEATURES
plt.plot(X)
```

```python
y= np.sin(X).ravel() #Target
plt.scatter(X,y) #Boşluklu şekilde grafiği oluşturur.
```

## Gürültü Ekleme

```python
y[::5] +=1*(0.5 - np.random.rand(8))
plt.scatter(X,y)
```

## Test Verisi Oluşturma: 0-5 arasında 500 adet sayı.

```python
T = np.linspace(0, 5, 500)[:, np.newaxis]
```

## Modelin Eğitilmesi ve Görselleştirilmesi

```python
for i,weight in enumerate(["uniform","distance"]): #KNN regresörünü iki farklı ağırlık stratejisi ile denemek için değişkenlerle döngü başlattık.
    knn = KNeighborsRegressor( n_neighbors=5, weights=weight ) # 5 komşulu modeli uniform=eşit ve distance=uzaklık ağırlık tiplerine göre model belirlenir.
    y_pred=knn.fit( X,y ).predict(T) #Model X,y ile eğitilir ve T test verisinde tahmin yapılır.
    plt.subplot(2,1,i+1) #Grafik oluşturma.
    plt.scatter( X, y, color="green", label="data" ) #Gerçek veriler yeşil olarak belirlenir.
    plt.plot( T, y_pred, color="blue", label="prediction" ) #Tahmin değerleri mavi olarak belirlenir.
    plt.axis("tight") #Grafik sınırları veriye göre düzenlenir.
    plt.legend() #Etiketleri gösterir.
    plt.title("KNN Regressor Weights = {}".format(weight))
plt.tight_layout()
plt.show()
```

#### OCR (Optical Character Recognition), yani Optik Karakter Tanıma, basılı ya da el yazısı metinleri dijital verilere çeviren teknolojidir.

#### Bu teknolojiden yararlanmak için Python programlama dilinde Tesseract kütüphanesinden faydalanılır.
